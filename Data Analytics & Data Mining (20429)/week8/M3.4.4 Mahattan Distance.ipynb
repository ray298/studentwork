#**Manhattan Distance**

The Manhattan distance get's its name form the New York city taxi drivers problem. The goal is  to calculate the shortest distance between 2 points. With the euclidean distance the shortest distance will be that calculated by pythagoras or the hypotonus of a triangle (the green line in Figure 1). But obviously if your a taxi driver you cannot drive through buildings so you have to do an up and across motion.

![alt text](https://www.computing.dcu.ie/~amccarren/mcm_images/Manhattan_distance.svg.png)

Figure 1: [Taxicab geometry](https://en.wikipedia.org/wiki/Taxicab_geometry)

So we now realise we cannot drive through buildings, which rules out the Euclidean distance as an appropriate formula:

>>$d(p,q)={\sqrt {\sum _{i=1}^{n}(q_{i}-p_{i})^{2}}}$

If we want to reflect the taxi drivers path then we should use the sides of each triangle. This  is effectively the Manhattan distance:

>>$d(p,q)=\sum_i^n\lvert (q_{i}-p_{i}) \rvert$

This Manhattan distance metric is also known as Manhattan length, rectilinear distance, L1 distance or L1 norm, city block distance, Minkowskiâ€™s L1 distance, taxi-cab metric, or city block distance. In some cases it can be preferable to use as a distance measure as it can put less weight on outlying points. It has also been found to be benifical when you have a very high dimensional dataset.

The code below is a very simple example of how it works using the scikit learn manhattan_distances library. Play about with it and compare it to the Euclidean distance. But your thoughts on the comments board.





```
from sklearn.metrics.pairwise import manhattan_distances
X = [[0, 1], [1, 2]]

print(manhattan_distances(X, X))
print(manhattan_distances(X, [[0, 0]]))
```
