#**Apriori Example**
The example we are going to use to demonstrate the apriori algorithm comes from [GeeksforGeeks](https://www.geeksforgeeks.org/implementing-apriori-algorithm-in-python/). In it we will pull down a online retail dataset and attempt to find the most frequent itemsets. I have concentrated the following code on the UK sales as these possess the most transactions. Now you will notice that when I am importing the data into google colabs that I have used "iso-8859-1" encoding. This was the one I found to work.

You should also look at how the Apriori algorithm needs to be set up. Each product must be set up as a hot encoded column for each invoice (basket). This [link](http://rasbt.github.io/mlxtend/api_subpackages/mlxtend.frequent_patterns/) should help you here.

In the final rules dataset you will see two columns ('antecedents','consequents') that might need some explaination. In the previous steps we refered to them as $X \mapsto Y$. $X$ is the antecedent and $Y$ is the consequent.
The remaining columns describe the metrics that we have explained in step step 4.1.4. Experiment with the data and see if it all makes sense.

Importing the relevant libraries

```
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

Reading the data. Note the encoding format, it is not our usual one.

```
dataset = [['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
```

Now we are encoding them so "mixtend" can take the data.

```
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
print(df)
from mlxtend.frequent_patterns import apriori
```

Now we will use apriori to generate the rules.

```
apriori(df, min_support=0.7,use_colnames=True)
```
