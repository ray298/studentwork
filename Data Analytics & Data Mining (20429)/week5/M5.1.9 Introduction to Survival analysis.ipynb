Survival Analysis is used to estimate the lifespan of a particular population under study. It is also called ‘Time to Event’ Analysis as the goal is to estimate the time for an individual or a group of individuals to experience an event of interest.This time estimate is the duration between birth and death events[1]

#Data#

A typical example of a survival analysis study would be a clinical trial looking at the impact of a drop on patients chances of survical over a period of time. The patients in the study don't all have to start at the same time but their starting points are all referred to as time 0 (t=0) and each subject has probabliity of 1 to survie time 0. This type of study is typically know as a longtitudal study. 


##Censorship##

It is important to understand that not every member of the population will experience the Event of Interest (death, churn, etc) during the study period. We do not know when they will experience the event of interest as of the time of the study. Their survival times are thus, labelled as ‘Censored’[1]. Therefore, Censorship allows you to measure lifetimes for the population who haven’t experienced the event of interest yet.

There are different types of Censorship done in Survival Analysis as explained below[3]. Note that Censoring must be independent of the future value of the hazard for that particular subject [5].

* Right Censoring: 

This happens when the subject enters at t=0 i.e at the start of the study and terminates before the event of interest occurs. 

* Left Censoring:

This happens when the birth event wasn’t observed.  The concept known as Length-Biased Sampling should  be mentioned here. This type of sampling occurs when the goal of the study is to perform analysis on the people/subjects who already experienced the event and we wish to see whether they will experience it again

* Interval Censoring:

This happens when the follow-up period, i.e time between observation, is not continuous. This can be weekly, monthly, quarterly, etc.

* Left Truncation

It is referred to as late entry. The subjects may have experienced the event of interest before entering the study.

##Survival Function##

we are now going to introduce the concept of a survival functiion and this can 
be given by

$S(t) = P(T>t)$

The Survival Function defines the probability that the event of interest has not occurred at time t. It can also be interpreted as the probability of survival after time t [7]. Here, T is the random lifetime taken from the population and it cannot be negative. Note that S(t) is between zero and one (inclusive), and S(t) is a non-increasing function of t,[3].

#Hazard Function

The Hazard Function also called the intensity function, is defined as the probability that the subject will experience an event of interest within a small time interval, provided that the individual has survived until the beginning of that interval [1].It can also be considered as the risk of experiencing the event of interest at time t. It is the number of subjects experiencing an event in the interval beginning at time t divided by the product of the number of subjects surviving at time t and interval width[2].

 This scales the equation by ΔT[4]. The equation of the Hazard Rate is given as:

\begin{equation}
h(t) = \lim_{\delta \to 0} \frac {Pr(t\le T\le t+\delta t \mid T \ge t)}{\delta t}
 \end{equation}

In Practice the $\delta t$ is a usually the next period of time.
