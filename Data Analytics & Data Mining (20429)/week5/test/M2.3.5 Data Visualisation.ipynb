# **Data Visualisation**

When we do most things in life we use our senses to understand and feel things. Our brain is absorbing an enormous amount of information, breaking it down or categroising it into groups in order to make decisions. Data visualistion is the process of making a picture from a vast amount of information. Remember a picture paints a thousand words, so try an decide how you will present your visualisations to your final audience.  It can give us a clarity we would not normally get when looking at a vast amount of data/words. For example, we have all looked at a spreadsheet or a database, and quite frankly got lost in the data. The following link is a TED talk [David McCandless](https://www.youtube.com/watch?v=5Zg-C8AAIGg) gave on data visualisation, its a little long (18 minutes) but it is worth it. It is incredibly interesting and shows the power of visualisation. 

I always tell students when completing their visualisations to be selective. Imagine you go to an art gallery or museum like the Lourve. We always enter these places with excitment (well I do), and having walked around for a number of hours we get desensitised and can really miss the story of a wonderful painting. This is the same with visualisations in analytics. Sometimes we can have to many pictures, so be selective and try and get to the point as quickly as possible.

This article by [Towards Science](https://towardsdatascience.com/introduction-to-data-visualization-in-python-89a54c97fbed) is a really nice introduction to visualisation in python. It works on the IRIS dataset and and quickly demonstrates differing charts. 

Before we go to the next step I would like you think about why we visualise data. We do it to tell a story but also to highlight characteristics such as anomlies or outliers. Ask yourself what an outlier is and what you should do with them.
