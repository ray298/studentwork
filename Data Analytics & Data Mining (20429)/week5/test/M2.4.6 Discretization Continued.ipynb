#**Discretization Contnued**

In topic M2.3.11 we discussed binning techniques that could be used to discretize a feature. These processes while useful in understanding and preprocessing our data can also be considered data numerosity reduction techniques, as we are attempting to classify continous variables.

Generally we have 3 types of attributes or features:

* Nominal Variables - The values from an unordered set
>* (Male, Female)
>* (Yes/No)
>* (Conservative/Labour/Liberal/SNP/Green Party)
* Ordinal Variables - The values from an ordered set
>* Very Unsatisfied – 1
>* Unsatisfied – 2
>* Neutral – 3
>* Satisfied – 4
>* Very Satisfied – 5

* Continous- These are real numbers and are often described as interval scales or ratio scales. Have a look at the following [link](https://www.questionpro.com/blog/ratio-scale/) for a more detailed explaination of ratio and intervalvariables.

As we said in Topic 3 of this MOOC Discritization allows us to divide a continous variable into intervals/categories.

Further to the techniques we have looked at in previous Topics in this MOOC we will look a very simple rule natural partitioning apporach known as the 3-4-5 rule and an entropy-based discretization method which takes account of the output variable when deciding on how to discretize an input variable.
