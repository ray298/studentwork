# **Heteroscedasticity and Linear Regression**
 
In earlier topics we discussed the assumptions when building regression models.
The simplest version of a linear regression model is $Y_i= a +b_1 +\epsilon_i$.
Where $\epsilon_i$ is randomly and independently normally distributed around 0 with a constand variance. Heteroscedasticity occures when the variation in the error terms varies is not constant. It can be caused by the following:

* when there is a large range between the largest and smallest observed values (outliers).
* When the model is not correctly specified 
* If observations are from mixed scales (farenheit to celcius) 
* Skewness  in the distribution of the regressor variable(dependent).

From the above list we can see that we are assuming a linear model, but by using a linear model and examining the errors we can expose potential outliers in our data. I am not saying we have to stick with this model but using it initially can help you identify outlying observations.
